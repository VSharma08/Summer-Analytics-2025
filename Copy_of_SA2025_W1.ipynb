### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
data['mpg'] = pd.to_numeric(data['mpg'], errors='coerce')
median_mpg = data['mpg'].median()

above_median_mpg_cars = data[data['mpg'] > median_mpg].copy()

above_median_mpg_cars['hp_to_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']

max_ratio = above_median_mpg_cars['hp_to_weight_ratio'].max()

best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp_to_weight_ratio'] == max_ratio]

print(best_ratio_cars['name'])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt
import seaborn as sns

data['mpg'] = pd.to_numeric(data['mpg'], errors='coerce')

avg_mpg_by_year_origin = data.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin')
plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
data_reset = data.reset_index()

data_reset['mpg'] = pd.to_numeric(data_reset['mpg'], errors='coerce')

# Car names that appear in more than one model_year
name_counts = data_reset.groupby('name')['model_year'].nunique()
multi_year_cars = name_counts[name_counts > 1].index

# Filtering the original dataframe to include only multi-year cars
multi_year_data = data_reset[data_reset['name'].isin(multi_year_cars)]

# Computing the standard deviation and mean of mpg for each car name across years
mpg_stats = multi_year_data.groupby('name')['mpg'].agg(['std', 'count', 'mean'])

# Identifying consistent cars: std(mpg) < 1.0
consistent_cars = mpg_stats[mpg_stats['std'] < 1.0]

# Finding the cars with the lowest standard deviation among consistent cars
if not consistent_cars.empty:
    min_std = consistent_cars['std'].min()
    lowest_variation_cars = consistent_cars[consistent_cars['std'] == min_std]

    lowest_variation_cars = lowest_variation_cars.rename(columns={
        'std': 'mpg_std_dev',
        'count': 'number_of_appearances',
        'mean': 'average_mpg'
    })

    result = lowest_variation_cars[['number_of_appearances', 'average_mpg']]

    result_sorted = result.sort_values(by=['number_of_appearances', 'average_mpg'], ascending=[False, False])

    print("Consistent car model(s) with the lowest variation in mpg:")
    print(result_sorted)
else:
    print("No consistent car models found meeting the criteria.")
```
